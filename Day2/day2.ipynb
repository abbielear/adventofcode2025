```python
with open('input.inp') as f:
    ranges_str = f.read()
ranges_list = ranges_str.split(',')
```

```python
invalid = []

for range_pair in ranges_list:
    start, end = [int(x) for x in range_pair.split('-')]
    #print(start, end)
    for test_number in range(start, end+1):
        test_number_str = str(test_number)
        #print(test_number_str)
        
        if len(test_number_str) % 2 != 0:
            continue
        halfway_index = len(test_number_str) // 2

        if test_number_str[:halfway_index] == test_number_str[halfway_index:]:
            #print(test_number_str, 'match')
            invalid.append(test_number)

print(sum(invalid))
```
