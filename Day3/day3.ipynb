```python
import numpy as np
```

```python
with open('input.inp') as f:
    batteries = f.readlines()
```

```python
batteries_arr = np.empty((len(batteries), len(batteries[0])-1), dtype='int')
for i, battery_str in enumerate(batteries):
    batteries_arr[i] = [int(char) for char in battery_str[:-1]]
```

```python
#Part1

total = 0

for battery in batteries_arr:
    maximum = np.max(battery[:-1])
    #print(battery[:-1])
    max_index = np.argmax(battery[:-1])
    #print(maximum,max_index)
    second_maximum = np.max(battery[max_index+1:])
    #print(maximum, second_maximum)
    joltage = int(f'{maximum}{second_maximum}')
    total += joltage

print(total)
```

```python
total = 0

for battery in batteries_arr:
    joltage_list = []
    previous_index = -1
    for jolt_index in range(12):
        end_index = 11 - jolt_index
        if end_index == 0:
            accessible_array = battery[previous_index+1:]
        else:
            accessible_array = battery[previous_index+1:-end_index]
        maximum = np.max(accessible_array)
        joltage_list.append(maximum)
        max_index = np.argmax(accessible_array) + previous_index+1
        previous_index = max_index

    joltage_str = "".join([str(i) for i in joltage_list])
    #print(joltage_str)
    total += int(joltage_str)

print(total)
```
