```python
with open('input.inp') as f:
    lines = f.readlines()
```

```python
#make sure the final line of the input also has a '/n'
```

```python
import numpy as np
```

```python
diagram_array = np.full((len(lines)+2, len(lines[0])+1), fill_value=0)

for line_index, line in enumerate(lines):
    diagram_array[line_index+1][1:] = [1 if char=='@' else 0 for char in line]

diagram_array
```

```python
def get_sum_around_pos(row_index, column_index):
    above =         diagram_array[row_index-1,  column_index]
    above_right =   diagram_array[row_index-1,  column_index+1]
    right =         diagram_array[row_index,    column_index+1]
    below_right =   diagram_array[row_index+1,  column_index+1]
    below =         diagram_array[row_index+1,  column_index]
    below_left =    diagram_array[row_index+1,  column_index-1]
    left =          diagram_array[row_index,    column_index-1]
    above_left =    diagram_array[row_index-1,  column_index-1]

    return sum([above, above_right, right, below_right, below, below_left, left, above_left])
```

```python
#Part1

total = 0

for row_index, row in enumerate(diagram_array):
    for column_index, value in enumerate(row):
        if value == 1:
            sum_around_pos = get_sum_around_pos(row_index, column_index)
            if sum_around_pos < 4:
                #print(row_index, column_index)
                total += 1

total
```

```python
#Part2

total = 0
updated_diagram = True

while updated_diagram == True:
    updated_diagram = False
    for row_index, row in enumerate(diagram_array):
        for column_index, value in enumerate(row):
            if value == 1:
                sum_around_pos = get_sum_around_pos(row_index, column_index)
                if sum_around_pos < 4:
                    total += 1
                    diagram_array[row_index,column_index] = 0
                    updated_diagram = True

total
```
