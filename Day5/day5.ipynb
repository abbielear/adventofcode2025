```python
with open('input.inp') as f:
    contents = f.read()
```

```python
fresh_ranges, IDs = contents.split('\n\n')
```

```python
fresh_count = 0

for ID_str in IDs.split('\n'):
    for fresh_range in fresh_ranges.split('\n'):
        first_index, last_index = fresh_range.split('-')
        if int(first_index) <= int(ID_str) <= int(last_index):
            #print(ID_str, 'fresh')
            fresh_count += 1
            break

fresh_count
```

```python
fresh_ranges_list = fresh_ranges.split('\n')

fresh_ranges_list = [[int(x.split('-')[0]), int(x.split('-')[1])] for x in fresh_ranges_list]
```

```python
#Part2

#previous_ranges = []
new_updates = True

while new_updates == True:
    new_updates = False
#for i in range(10):
    print('starting')
    for fresh_range_index, fresh_range in enumerate(fresh_ranges_list):
        print('selected', fresh_range)
        sel_first_index, sel_last_index = fresh_range

        for test_range_idx, test_range in enumerate(fresh_ranges_list[:fresh_range_index] + fresh_ranges_list[fresh_range_index+1:]):
            print('testing against', test_range)
            test_first_index, test_last_index = test_range
            #check if sel range inside test range
            if sel_first_index > test_first_index and sel_last_index < test_last_index:
                print('inside')
                fresh_ranges_list.remove([sel_first_index, sel_last_index])
                break
            #Check if sel range entirely outside test range
            if sel_first_index < test_first_index and sel_last_index > test_last_index:
                print('outside')
                fresh_ranges_list.remove([test_first_index, test_last_index])
                break
            
            if sel_first_index == test_first_index and sel_last_index == test_last_index:
                print('identical')
                fresh_ranges_list.remove([sel_first_index, sel_last_index])
                break

            #check if new lower value is lower than old and upper value inside or equal
            if sel_first_index < test_first_index and test_first_index <= sel_last_index <= test_last_index:
                fresh_ranges_list[test_range_idx][0] = sel_first_index
                fresh_ranges_list.remove([sel_first_index, sel_last_index])
                new_updates = True
                print('updating')
                break
            #check if new upper value is higher than old and lower value inside or equal
            if test_first_index <= sel_first_index <= test_last_index and sel_last_index > test_last_index:
                fresh_ranges_list[test_range_idx][1] = sel_last_index
                fresh_ranges_list.remove([sel_first_index, sel_last_index])
                new_updates = True
                print('updating')
                break

        print(fresh_ranges_list)

fresh_count = 0

for fresh_range in fresh_ranges_list:
    #print(len(range(previous_range[0], previous_range[1])))
    fresh_count += len(range(fresh_range[0], fresh_range[1]+1))

print(fresh_count)
```
